```python
import pandas as pd
class pau():
    df = pd.read_csv('sis.csv')
    mini = 0
    maxi = 0
    mat = []
    name = []
    age = []
    grade = []
    def __init__(self):
        for i in range(len(self.df.Age)):
            if self.mini < self.df.Age[i] < self.maxi:
                self.mat.append(self.df.MatNo[i])
                self.name.append(self.df.Name[i])
                self.age.append(self.df.Age[i])
                self.grade.append(self.df.Grade[i])
        self.stuff = pd.DataFrame({"MatNo": self.mat, "Name": self.name, "Age": self.age, "Grade": self.grade})
    def show(self):
        self.stuff.to_csv(self.nom)
        return(self.stuff)
        
class The_Pirates(pau):
    mini = 14
    nom = "The_Pirates.csv"
    maxi = 18
    def __init__(self):
        pau.__init__(self)
        
class The_Yankees(pau):
    mini = 18
    maxi = 22
    nom = "The_Yankees.csv"
    def __init__(self):
        pau.__init__(self)
    
class The_Bulls(pau):
    mini = 22
    maxi = 100000
    nom = "The_Bulls.csv"
    def __init__(self):
        pau.__init__(self)
        
pi = The_Pirates()
pi.show()
```
